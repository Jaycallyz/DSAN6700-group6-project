```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

book = pd.read_csv("data/clean_data/book_processed.csv")
movie = pd.read_csv("data/clean_data/movie_processed.csv")
music = pd.read_csv("data/clean_data/music_processed.csv")
dest = pd.read_csv("data/clean_data/destination_processed.csv")

```

```python
book["text_len"] = book["text_representation"].astype(str).str.len()
movie["text_len"] = movie["text_representation"].astype(str).str.len()
music["text_len"] = music["text_representation"].astype(str).str.len()
dest["text_len"] = dest["text_representation"].astype(str).str.len()
```

```python
df_len = pd.DataFrame({
    "Books": book["text_len"],
    "Movies": movie["text_len"],
    "Music": music["text_len"],
    "Destinations": dest["text_len"]
})

plt.figure(figsize=(10,6))
sns.boxplot(data=df_len)
plt.title("Text Length Distribution Across Modalities")
plt.ylabel("Number of Characters")
plt.xlabel("Modality")
plt.show()
```

```python
counts = {
    "Books": len(book),
    "Movies": len(movie),
    "Music": len(music),
    "Destinations": len(dest)
}

plt.figure(figsize=(8,5))
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Dataset Size by Modality")
plt.ylabel("Number of Items")
plt.xlabel("Modality")
plt.show()
```

```python
top_countries = dest["country"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Countries by Number of Destinations")
plt.ylabel("Count")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
```
